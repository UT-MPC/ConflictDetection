```python
from alpha_test import test_auc
import logging
import numpy as np
logging.basicConfig(level=logging.INFO)
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from config import *
from utils import *
from ContextAccessor import ContextAccessor
ctx_info = ContextAccessor({
            # TIME_CTX: {
            #     "range" : (0, 24*60),
            #     "interval" : 20,
            # },
            # "humidity#NUM" : {
            #     "range" : (0., 1.0),
            #     "interval" : 0.1,
            # },
            # "apparentTemperature#NUM" : {
            #     "range" : (-10., 100),
            #     "interval" : 10,
            # },
            "summary#CAT": {},
            WEEKDAY_CTX: {
                "range": (0, 6),
                "interval": 1,
            },
        })
```

```python

def find_optimal_alpha(alpha_range):
    auc_values = []
    for alpha in alpha_range:
        auc_values.append((test_auc(alpha), alpha))
        print(auc_values[-1])
    return auc_values

def find_optimal_alpha_nonLinear(alpha_min, alpha_max, alpha_step):
    auc_values = []
    alpha = alpha_min
    while alpha <= alpha_max:
        auc, _ = test_auc(alpha, ctx_info)
        auc_values.append((auc,alpha))
        print(auc_values[-1])
        if alpha_step <= alpha / 10:
            alpha_step *= 10
        alpha += alpha_step
    return auc_values
print("Run test AUC for context " + str(ctx_info.get_all_ctx_ordered()))

auc_values = find_optimal_alpha_nonLinear(5e-6, 1.0, 1e-6)
# print(auc_values)
print("max auc is " + str(max(auc_values)))
```
